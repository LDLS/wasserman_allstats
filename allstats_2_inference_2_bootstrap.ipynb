**8.1.**  Find the plug-in estimate of the correlation coefficient of the following data.

```python
LSAT = np.array([576, 635, 558, 578, 666, 580, 555, 661, 651, 605, 653, 575, 545, 572, 594])
GPA = np.array([3.39, 3.30, 2.81, 3.03, 3.44, 3.07, 3.00, 3.43, 3.36, 3.13, 3.12, 2.74, 2.76, 2.88, 2.96])
```

Estimate the standard error using the bootstrap. Find a 95 percent confidence interval using the Normal, pivotal, and percentile methods.

```python
%%time
tuples = np.hstack([LSAT.reshape(-1,1), GPA.reshape(-1,1)])
corr_plugin = corrcoeff(LSAT,GPA)

n = tuples.shape[0]
B = 1000

idx = np.random.choice(tuples.shape[0], size = (B, n), replace = True)

bootstrap_tuples = tuples[idx]
LSAT_B = bootstrap_tuples[:,:,0]
GPA_B = bootstrap_tuples[:,:,1]


corr_boot = boot_corrcoeff(LSAT_B, GPA_B)
seboot = corr_boot.std()

ia, ib = np.percentile(corr_boot, [2.5, 97.5])

print(corr_plugin)
print('bootstrap normal confidence iterval: ', [corr_plugin - z_5pc*seboot, corr_plugin + z_5pc*seboot])
print('bootstrap percentile confidence iterval: ', [ia, ib])
print('bootstrap pivotal confidence iterval: ', [2*corr_plugin - ib, 2*corr_plugin - ia])
```

```python
fig = plt.figure(figsize = (6,4))
plt.hist(corr_boot, bins = 15)
plt.tick_params(axis = 'x', labelsize = 10)
plt.tick_params(axis = 'y', labelsize = 10);
```

**8.2** Conduct a simulation to compare the various bootstrap confidence interval methods. Let $n = $ 50 and let $T(F) =
\int (x -\mu)^3 dF(x)/ \sigma^3$ be the skewness. Draw $Y_1, \ldots ,Y_n \sim N(0,1)$ and set $X_i = e^{Y_i}$, $i = 1, \ldots,n$. Construct the three types of bootstrap 95 percent intervals for $T(F)$ from the data $X_1, \ldots, X_n$. Repeat this whole
thing many times and estimate the true coverage of the three intervals.

A common parametrization for a lognormal random variable $Y$ is in terms of the mean, $\mu$, and standard deviation, $\sigma$, of the unique normally distributed random variable $X$ such that $e^X = Y$. This parametrization corresponds to setting `s`$ = \sigma$ and `scale` $ = e^\mu$.

```python
%%time

nsamples = 10**2
min_nsamples = 10

mean = 0.
sigma = 1.

bootstrap_reps = 10**3
nexps = 10**2

#### Draw data with the appropiate distribution
data = np.random.lognormal(mean = mean, sigma = sigma, size = nsamples)

### Get ECDF and CDF
ecdf_data = ECDF(data)
cdf_data = sp.stats.lognorm.cdf(ecdf_data.x, s = sigma, scale = np.exp(mean))

distro_plots(data, ecdf_data, cdf_data = cdf_data)
```

```python
%%time
#### Experiment parameters
exp_sample_sizes, exp_slices = exp_params(nsamples, min_nsamples)

#### Draw boostrap samples
bootstrap_data = np.random.choice(data, (nsamples, bootstrap_reps, nexps), replace = True)

#### Compute true statistic, plugin estimate and bootstrap estimate trials
statistic_name = 'skewness'

true_statistic = sp.stats.lognorm.stats(loc = mean, s = sigma, moments = 's').tolist() 
# equivalently (np.exp(sigma**2) + 2)*np.sqrt(np.exp(sigma**2) - 1)

plugin_estimate = exp_skew(data, slices = exp_slices)

bootstrap_estimate = exp_skew(exp_data = bootstrap_data, slices = exp_slices)

### Obtain experiment intervals
bootstrap_intervals, bootstrap_coverage = exp_intervals(exp_sample_sizes, 
                                                        bootstrap_estimate, 
                                                        plugin_estimate, 
                                                        true_statistic)

print('Statistic true value:', true_statistic)
print('Plug in estimate:', plugin_estimate[-1])
```

**8.3**  Let $X_1, \ldots, X_n \sim t_3$ where $n = 25$. Let $\theta = T(F) = (q_{.75} - q_{.25})/1.34$ where $q_p$ denotes the
$p^{th}$ quantile. Do a simulation to compare the coverage and length of the following confidence intervals for $\theta$: (i) Normal interval with standard error from the bootstrap, (ii) bootstrap percentile interval, and (iii) pivotal bootstrap interval.

Recall $X$ has (Student's) $t$-distribution with $\nu$ degrees of freedom if is PDF is given by

$$
f_X(t) = \frac{\Gamma\bigl(\frac{\nu + 1}{2}\bigr)}{\sqrt{\nu\pi}\Gamma\bigl(\frac{\nu}{2}\bigr)} \Bigl( 1 + \frac{t^2}{\nu} \Bigr)^{-\frac{\nu +1}{2}}
$$

For $Y_1, \ldots, Y_n \sim N(\mu, \sigma^2)$ IID, if $S^2 = \frac1{n-1} \sum_{n = 1}^n (Y_i - \overline{Y}_n)^2$ is the Bessel-corrected (unbiased) sample variance, then
$$
\frac{\overline{Y}_n - \mu}{\sigma/\sqrt{n}} \sim N(0,1)
$$
and
$$
\frac{\overline{Y}_n - \mu}{S/\sqrt{n}} \sim t_{n-1} \to N(0,1) \; \textrm{as} \; n \to \infty
$$

```python
%%time

nsamples = 100
min_nsamples = 10

nu = 3

bootstrap_reps = 10**3
nexps = 10**2

#### Draw data with the appropiate distribution
data = np.random.standard_t(nu, nsamples)

### Get ECDF and CDF
ecdf_data = ECDF(data)
cdf_data = sp.stats.t.cdf(ecdf_data.x, df = nu)

distro_plots(data, ecdf_data, cdf_data = cdf_data)
```

**8.7** Let $X_1,\ldots, X_n \sim U(0, \theta)$. Let $\hat{\theta}_n = \mathrm{max}\{X_1, \ldots, X_n\}$. Generate a dataset of size $50$ with $\theta = 1$. (i) Find the distribution of $\hat{\theta}_n$ and compare it to the histograms from the bootstrap. (ii) This is a case where the bootstrap does very poorly. In fact, we can prove this is the case. Show that $P(\hat{\theta}_n = \theta) = 0$ and yet $P(\hat{\theta}^* = \hat{\theta}) \approx 0.632$. Hint: show that $P(\hat{\theta}^* = \hat{\theta})= 1 - (1 - (1/n))^n$.

Let $X_1,\ldots, X_n \sim U(0, \theta)$ and $\hat{\theta}_n = \mathrm{max}\{X_1, \ldots, X_n\}$. Then,
$$
P(\hat{\theta}_n \leq x) 
= \prod_{i = 1}^n P(X_i \leq x)
= P(X \leq x)^n
= \frac{x^n}{\theta^n} \, 1_{[0,\theta]}(x) + 1_{(\theta, \infty)}(x).
$$
Hence,
$$
f_{\hat{\theta}_n}(x) = \frac{n\,x^{n-1}}{\theta^n} 1_{[0,\theta]}(x)
$$
and
$$
P(\hat{\theta}_n = \theta) 
= \lim_{\varepsilon \to 0} \frac{n}{\theta^n} \int_{\theta - \varepsilon}^\theta x^{n-1} d\,x
= 0
$$

Now, since $X^*_i|(X_1, \ldots X_n)\sim F_n$, with $F_n$ the ECDF for $i = 1, \ldots, n$,
$$
P(\hat{\theta}^* \leq x) 
= \prod_{i = 1}^n P(X^*_i \leq x)
= F_n(x)^n
= \Biggl( \frac{1}{n} \sum_{i=1}^n 1_{ \{X_i \leq x\}}(x) \Biggr)^n.
$$
Thus, $P(\hat{\theta}^* \leq \theta)  = 1$ and 
\begin{align*}
P(\hat{\theta}^* = \theta) 
&= 1 - P(\hat{\theta}^* < \theta) \\
&= 1 - \prod_{i = 1}^n P(X^*_i < \theta) \\
&= 1 - P(X^* < \theta)^n \\
&= 1 - \bigl( P(X^* \leq \theta) - P(X^* = \theta)  \bigr)^n \\
&= 1 - \biggl( 1 - \frac{1}{n}  \biggr)^n \underset{n \to \infty}{\longrightarrow} 1 - e^{-1} \sim 0.6321205588285577
\end{align*}


```python
%%time

nsamples = 10**3
min_nsamples = 10**2

theta = 1.

bootstrap_reps = 10**3
nexps = 10**2

#### Draw data with the appropiate distribution
data = np.random.uniform(low = 0, high = theta, size = nsamples)

### Get ECDF and CDF
ecdf_data = ECDF(data)
cdf_data = sp.stats.uniform.cdf(ecdf_data.x, loc = 0, scale = theta)

distro_plots(data, ecdf_data, cdf_data = cdf_data)
```
